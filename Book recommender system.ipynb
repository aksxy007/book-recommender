## Collaborative Filltering Based Book Recommender Sytem

```python
x=ratings_with_book_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
```

```python
ratings_with_book_name['User-ID'].isin(padhe_likhe_users)
```

```python
filtered_rating=ratings_with_book_name[ratings_with_book_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()["Book-Rating"]>=50
famous_books=y[y].index
```

```python
final_Ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_Ratings.drop_duplicates()
```

```python
pt=final_Ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores =cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #index_fetch
    book_index=np.where(pt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_scores[book_index])),key=lambda x:x[1],reverse=True)[1:11]
    for i in similar_books:
        print(pt.index[i[0]])
#     return suggestions
```

```python
recommend('Message in a Bottle')
```
